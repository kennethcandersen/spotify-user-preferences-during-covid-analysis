```python
# Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import json
import re, glob
import os, sys
from scipy import stats
```

```python
country_tracks_data_df = pd.read_csv("New Zealand_FIX_tracks_with_audio_features.csv")

country_tracks_data_df.head()
```

```python
cor_country_tracks_data_df = country_tracks_data_df.rename(columns={"Energy": "Energy2", 'Valence': "Energy"})
cor_country_tracks_data_df.head()
```

```python
cor_country_tracks_data_df = cor_country_tracks_data_df.rename(columns={"Energy2": "Valence"})
cor_country_tracks_data_df.head()
```

```python
Tot_Streams = country_tracks_data_df['Streams'].sum() 
Tot_Streams 
cor_country_tracks_data_df['Stream %'] = cor_country_tracks_data_df['Streams'] / Tot_Streams
cor_country_tracks_data_df.head()
```

```python
cor_country_tracks_data_df['Danceability_Stream%'] = cor_country_tracks_data_df['Danceability'] * cor_country_tracks_data_df['Stream %']    
cor_country_tracks_data_df['Valence_Stream%'] = cor_country_tracks_data_df['Valence'] * cor_country_tracks_data_df['Stream %'] 
cor_country_tracks_data_df['Energy_Stream%'] = cor_country_tracks_data_df['Energy'] * cor_country_tracks_data_df['Stream %'] 
cor_country_tracks_data_df.head()
```

```python
cor_country_tracks_data_df.to_csv("New Zealand_tracks_with_audio_features.csv")
```
